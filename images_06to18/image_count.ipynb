```python
import pandas as pd

metadata = pd.read_csv('train_crop_images.csv', 
                 na_values=['NA', 'N/A', ''],
                 keep_default_na=True,
                 quotechar='"',
                 doublequote=True,
                 engine='python')


crop_classes = ["B11", "B12", "B13", "B14", "B15", "B16", "B21", "B22", "B31", "B32", "B33", "B55"] 
selected_metadata = metadata[metadata['lc1'].isin(crop_classes)]

class_counts = selected_metadata['lc1'].value_counts()
print(class_counts)

n_min_crop = class_counts.sort_values()[0]
print (n_min_crop)
```
