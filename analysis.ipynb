```python
import os
"""Counts images for each subfolder in train/test set
    To verify that no errors occured and all classes have equal amounts of images"""

def count_images_per_class(mainfolder):
    print("Images per folder:")
    print("-" * 50)
    
    # For one level of subfolders
    for eintrag in os.listdir(mainfolder):
        folder_path = os.path.join(mainfolder, eintrag)
        
        # Only folders, not files
        if os.path.isdir(folder_path):
            jpg_files = [f for f in os.listdir(folder_path) 
                          if f.lower().endswith('.jpg')]
            count = len(jpg_files)
            
            print(f"{eintrag:30} → {count:4} .jpg Dateien")
    
    print("-" * 50)


mainfolder = "./crop_images_balanced/train"
count_images_per_class(mainfolder)
```
